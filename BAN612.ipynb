```python
from datetime import datetime, timedelta

result['Posted Date'] = pd.to_datetime(result['Posted Date'])

current_date = datetime.now()

date_ranges = [
    (current_date - timedelta(days=70), current_date - timedelta(days=61)),  # 61-70 days ago
    (current_date - timedelta(days=60), current_date - timedelta(days=51)),  # 51-60 days ago
    (current_date - timedelta(days=50), current_date - timedelta(days=41)),  # 41-50 days ago
    (current_date - timedelta(days=40), current_date - timedelta(days=31)),  # 31-40 days ago
    (current_date - timedelta(days=30), current_date - timedelta(days=21)),  # 21-30 days ago
    (current_date - timedelta(days=20), current_date - timedelta(days=11)),  # 11-20 days ago
    (current_date - timedelta(days=10), current_date),   # 0-10 days ago
]

labels = [
    '61-70 days ago',
    '51-60 days ago',
    '41-50 days ago',
    '31-40 days ago',
    '21-30 days ago',
    '11-20 days ago',
    '0-10 days ago',
]

result['Posted Date Category'] = pd.cut(result['Posted Date'], bins=[date_range[0] for date_range in date_ranges] + [current_date], labels=labels)

job_counts = result['Posted Date Category'].value_counts()
for category, count in job_counts.items():
    print(f'Number of jobs posted in the last {category}: {count}')
```
